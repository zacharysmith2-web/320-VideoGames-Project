```python
numerical_cols = Data.select_dtypes(include=np.number).columns.tolist()

for column in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=Data[column])
    plt.title(f'Boxplot for {column}')
    plt.xlabel(column)
    plt.show()

def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

if 'total' in numerical_cols:
    outliers = find_outliers_iqr(Data, 'total')
    print("--- Outliers in 'total' column ---")
    print(outliers)
else:
    print("--- The 'total' column does not exist or is not numerical ---")
    #This chunck of code creates a box plot to help us find out if there are any outliers that might cause our data with
    #the ratings to be higher or lower than intended because of the outliers.
```

# Principles of DM

Some Principles of DM are Data Quality, which is to maintain accuracy, completeness and consistency which makes high quality data very valuable, which is important since without Data Quality checks you can be given a low quality data set and have it be basically useless because nothing note worthy can be done with it since it is low quality. Another principle is Data Security, which is to protect data from unauthorized access, breaches, and misuse, which is important because if you are working with data that involves sensitive information like people's bank details or something like that without that security that data can be out in the open and be used by anyone for malicious activity. Another principle is Data Ethics, which is to adhere to ethical standards in collecting. processing, and analysis of data, which is important since if you are not ethically sourcing or data or using data ethically it could harm people and ruin the reputation for the company.

# Principles of FM

Some principles of FM are to keep the original version, which is important since if you modify anything with the data and do not keep the original then if something goes wrong with the modified data set you do not have a backup to recover from and it will be hard to see how or why the changes to the data set happened if you delete the original. Another principle is to place comments and notes, which is important so that people can see what or why a line of code does and is used for. This is also helpful for the collecting of data to show where the data is sourced and what each of the columns of the Data Set means. Another principle is back up the files, which is very important since if you do not back up the files and you have a system error or crash and cannot recover the original file then you know that you do not need to worry since you have a backup of the file and can go from the backup to get back on track with everything that happened before the system error or crash.
